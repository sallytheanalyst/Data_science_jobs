## Univariate analysis

```python
# Analysis on experience level
plt.figure(figsize=(4,4))

el= sns.countplot(data=df, x="experience_level", order= df["experience_level"].value_counts(ascending = False).index)
values = df['experience_level'].value_counts(ascending=False).values
el.bar_label(container=el.containers[0], labels=values);
plt.title('Experience level analysis')
plt.show();
```

#### It can be noticed from this analysis that Senior level/Expert had the highest count (280) and the least was the Executive level/Director(26)

```python
# Analysis on job titles
plt.figure(figsize=(10,10))

jt= sns.countplot(data=df, y="job_title", order= df["job_title"].value_counts(ascending = False).index)
values = df['job_title'].value_counts(ascending=False).values
jt.bar_label(container=jt.containers[0], labels=values);
plt.title('Job title analysis')

plt.show();
```

#### Data scientist job title had the highest count(143), next to data engineer(132) then data analyst(97)

```python
# Analysis on job titles, filtering to top 5
plt.figure(figsize=(7,3))

tt=df['job_title'].value_counts().head(5)
tp= sns.countplot(data=df, y=df['job_title'], order= tt.index)
values = df['job_title'].value_counts(ascending=False).values
tp.bar_label(container=tp.containers[0], labels=tt);
plt.title('Top 5 Job titles')

plt.show();
```

#### This shows the top 5 job roles and the highest being Data Scientist(143), the lowest being Research Scientist(16)

```python
# Analysis on employment type

plt.figure(figsize=(4,4))

et=sns.countplot(data=df, x='employment_type', order= df['employment_type'].value_counts(ascending=False).index)
values = df['employment_type'].value_counts(ascending=False).values
et.bar_label(container=et.containers[0], labels=values);
plt.title('Employment Type analysis')

plt.show();
```

#### 588 of workers are full-time staffs, 10 are part-time staffs, 5 are Contract staffs and 4 are freelancers.

```python
# Analysis on employee residence

plt.figure(figsize=(7,3))

ert=df['employee_residence'].value_counts().head(10)
er= sns.countplot(data=df, y=df['employee_residence'], order= ert.index)
values = df['employee_residence'].value_counts(ascending=False).values
er.bar_label(container=er.containers[0], labels=ert);
plt.title('Top 10 Employee Residence')
plt.show();
```

#### The US has the highest employee residence with 332 people, followed by Great Britain with 44, then India and Canada. The least country among this is Pakistan with 6 people residing in that country

```python
# Analysis on company location

plt.figure(figsize=(7,3))

clt=df['company_location'].value_counts().head(10)
cl= sns.countplot(data=df, y=df['company_location'], order= clt.index)
values = df['company_location'].value_counts(ascending=False).values
cl.bar_label(container=cl.containers[0], labels=clt);
plt.title('Top 10 Company Locations')
plt.show();
```

#### The US has the most company locations with 355, followed by Great Britain with 47, then India and Canada. The least country among this is Neatherland with 4 companies. 

```python
# Analysis on company size

fig,ax = plt.subplots(figsize=(4,4))
count= Counter(df['company_size'])
ax.pie(count.values(), labels=count.keys(), autopct=lambda p:f'{p:.2f}%')
ax.set_title("Percentage distribution of company size")
plt.show();
```

#### Medium sized companies are the largest with 53.71%, followed by large organizations with 32.62% then small sized organizations with 13.67%

```python
# Work-year analysis
plt.figure(figsize=(4,4))

wy=sns.countplot(data=df, x='work_year', order= df['work_year'].value_counts(ascending=False).index)
values = df['work_year'].value_counts(ascending=False).values
wy.bar_label(container=wy.containers[0], labels=values);
plt.title('Work-Year Analysis')

plt.show();
```

#### The year with more salaries paid is 2022 (318), then 2021 (217), the least is 2020 (72). We could attribute this to more people transitioning to the tech space also

```python
# Creating a new salary category

def salary_category(x):
    if x <= 80000:
        return "Low earner"
    elif x <= 160000:
        return "Average earner"
    elif x <= 320000:
        return "High earner"
    else:
        return "Expert earner"
        
df['salary_category']= df['salary_in_usd'].apply(salary_category)
df.head(2)
```

```python
# Analysis on salary in usd

plt.figure(figsize=(6,4))

ss=sns.countplot(data=df, x='salary_category', palette='magma_r', 
                 order= df['salary_category'].value_counts(ascending=False).index)
values = df['salary_category'].value_counts(ascending=False).values
ss.bar_label(container=ss.containers[0], labels=values);
plt.title('Salary analysis')

plt.show();
```

#### Those earning between 160,001 and 300,000 are the highest earned salary(263), while those earning above 320,000 are the least(10)

```python
# Creating a new remote ratio category

def remote_type(x):
    if x <= 0:
        return "On-site"
    elif x <= 50:
        return "Hybrid"
    else:
        return "Fully remote"
        
df['remote_type']= df['remote_ratio'].apply(remote_type)
df.head(2)
```

```python
# Analysis on remote ratio

plt.figure(figsize=(6,4))
rr=sns.countplot(data=df, x='remote_type', palette='magma', order= df['remote_type'].value_counts(ascending=False).index)
values = df['remote_type'].value_counts(ascending=False).values
rr.bar_label(container=rr.containers[0], labels=values);
plt.title('Remote Ratio Analysis')

plt.show();
```

#### 381 are working remotely, 127 are majorly on-site while 99 are hybrid workers

## Bivariate Analysis

```python
# Remote Ratio by work year

plt.figure(figsize=(4,4))
rw= sns.countplot(data=df, x='work_year', hue='remote_type',palette='magma',
                  order= df['work_year'].value_counts(ascending=False).index)
values = df['work_year'].value_counts(ascending=False).values
plt.title('Remote_ratio by work year')
plt.show();
```

#### Irrespective of the year, fully remote jobs have been the highest. However, on-site jobs were lower in 2020 but has gradually increased over the years, which could be due to the suppression of covid.

## Experience level analysis

```python
# Employment type by experience level

plt.figure(figsize=(4,4))

(pd.crosstab(df['employment_type'],df['experience_level'], 
             normalize='index')
   .plot.bar(stacked=True)
)
plt.title('Employment Type by Experience Level')
plt.show();
```

#### For Contract employment, there are more entry-level. For freelencers, Intermediate experienced level is most common. For full-time employment, Senior experience level is most common and for Part-time employment, Entry-level is most common

```python
# Analysis on top 3 job titles by experience level
plt.figure(figsize=(6,4))

ttl=df['job_title'].value_counts().head(3)
tpl= sns.countplot(data=df, x=df['job_title'], hue= 'experience_level', order= ttl.index)
values = df['job_title'].value_counts(ascending=False).values
plt.title('Top 3 Job titles vs Experience level')

plt.show();
```

#### Senior level of experience is the most common to the 3 job roles, followed by intermediate levels of experience

```python
# Analysis on company size by experience level
plt.figure(figsize=(6,4))

cse=sns.countplot(data=df, x=df['company_size'], hue= 'experience_level',
                   order= df['company_size'].value_counts(ascending=False).index)
values = df['company_size'].value_counts(ascending=False).values
plt.title('Company Size by Experience level')

plt.show();
```

#### The medium sized companies have higher senior experienced staffs compared to intermediate experienced staffs. Large sized and smaller companies have higher intermediate staffs to senior experienced staffs although small organizations have high entry level roles

```python
# Company location analysis
plt.figure(figsize=(5,4))
cla=df['company_location'].value_counts().head(5)
cls= sns.countplot(data=df, x=df['company_location'], hue= 'company_size', order= cla.index)
values = df['company_location'].value_counts(ascending=False).values
plt.title('Top 5 Company Location by Company Size')

plt.show();
```

#### Medium sized companies are the most common in the top 3 countries then large sized before small sized organizations but in the 4th and 5th countries Large sized organizations are the most common then medium sized to small sized

```python
# Company location analysis vs experience level
plt.figure(figsize=(5,4))

cle=df['company_location'].value_counts().head(5)
clea= sns.countplot(data=df, x=df['company_location'], hue= 'experience_level', order= cle.index)
values = df['company_location'].value_counts(ascending=False).values
plt.title('Top 5 Company Location by Experience Level')

plt.show();
```

#### Based on location, Senior/Expert level are the highest in US and Canada, Mid/Intermediate level are the highest in Great Britain and India while Entry-level is the highest in Denmark

```python
# Highest salary by job title

plt.figure(figsize=(5,5))

ts=df['job_title'].value_counts().head(15)
jt=sns.jointplot(data=df, y='job_title', x='salary_in_usd',hue='salary_category', ylim=35, kind='hist')
values = df['salary_in_usd'].value_counts(ascending=False).values
plt.title('Salary Category by job title')

plt.show();
```

#### The job role with the highest salary is Principal Data Engineer with almost 600,000

```python
#Calculating the average salary by company location

plt.figure(figsize=(5,4))

csa=df.groupby(['company_location'])['salary_in_usd'].mean()
kl= csa.sort_values(ascending=False).index
top_10_locations = kl[:10]
clv = top_10_locations.value_counts().head(10)
clsv = sns.barplot(data=df, y='company_location', x= 'salary_in_usd', order=clv.index)
values = kl.value_counts(ascending=False).values
for container in clsv.containers:
    clsv.bar_label(container)
plt.title('Average Salary For The Top 10 Company Locations')

plt.show();
```

#### This shows the Russia has the highest paying average salary of 157500 and the least 10th country is Canada with 99823.7 average salary
